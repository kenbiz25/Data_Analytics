```python

import tweepy
import pandas as pd

# Use Twitter API v2 Bearer Token (not Consumer/Access Keys)
BEARER_TOKEN = "token_here"

# Authenticate with Tweepy v2
client = tweepy.Client(bearer_token=BEARER_TOKEN)

search_query = "hiring SEO Specialist -is:retweet -is:reply -has:links"
no_of_tweets = 10

try:
    # Fetch tweets
    tweets = client.search_recent_tweets(query=search_query, max_results=no_of_tweets, tweet_fields=["created_at", "public_metrics", "source"], user_fields=["name"], expansions="author_id")

    # Extract user data
    users = {u["id"]: u["name"] for u in tweets.includes["users"]}

    # Convert tweets to DataFrame
    attributes_container = [[users[tweet.author_id], tweet.created_at, tweet.public_metrics["like_count"], tweet.source, tweet.text] for tweet in tweets.data]

    # Column names
    columns = ["User", "Date Created", "Number of Likes", "Source of Tweet", "Tweet"]

    # Create DataFrame
    tweets_df = pd.DataFrame(attributes_container, columns=columns)

    print(tweets_df)
except Exception as e:
    print("Status Failed On,", str(e))
```
